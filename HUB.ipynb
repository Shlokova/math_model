Вычисление геодезического расстояния

```python
R = 6371.3
```

```python
def geodetic_range(coord_1, coord_2):
    
    
    a1 = coord_1[0] * math.pi / 180
    a2 = coord_2[0] * math.pi / 180
    b1 = coord_1[1] * math.pi / 180
    b2 = coord_2[1] * math.pi / 180
    
    x1 = R * math.cos(a1) * math.cos(b1)
    y1 = R * math.cos(a1) * math.sin(b1)
    z1 = R * math.sin(a1)
    
    x2 = R * math.cos(a2) * math.cos(b2);
    y2 = R * math.cos(a2) * math.sin(b2);
    z2 = R * math.sin(a2);
    
    dist = pow(pow((x2 - x1), 2) + pow((y2 - y1), 2) + pow((z2 - z1), 2), 1 / 2)
    alfa = 2 * math.asin(dist / (2 * R))

    return alfa
```

Целевая функция:

```python
def func(coord_h):
    
    sum = 0
    for i in range(0, 71):
        sum += math.exp(geodetic_range(eu_df['airport coordinates'][i], coord_h)) * eu_df['passenger traffic'][i]
    
    for i in range(0, 50):
        sum += math.exp(geodetic_range(aus_df['airport coordinates'][i], coord_h) * aus_df['passenger traffic'][i])
    
    return sum
```

Метод постоянного шага для одномерной оптимизации:

```python
def minimization(f, coord_h, h, flag):
    h = 180 * h  / (math.pi * R)
    if flag:
        if f([coord_h[0] + h, coord_h[1]]) < f(coord_h):
            while f([coord_h[0] + h, coord_h[1]]) < f(coord_h):
                coord_h[0] += h
                #print("Значение координат: ", coord_h[0], coord_h[1],"Значение целевой функции: ",  f([coord_h[0], coord_h[1]]))
        elif f([coord_h[0] - h, coord_h[1]]) < f(coord_h):
            while f([coord_h[0] - h, coord_h[1]]) < f(coord_h):
                coord_h[0] -= h
                
        print("Значение координат: ", coord_h[0], coord_h[1],"Значение целевой функции: ",  f([coord_h[0], coord_h[1]]))
        return coord_h[0]
    else :
        if f([coord_h[0], coord_h[1] + h]) < f(coord_h):
            while f([coord_h[0], coord_h[1] + h]) < f(coord_h):
                coord_h[1] += h
                #print("Значение координат: ", coord_h[0], coord_h[1],"Значение целевой функции: ",  f([coord_h[0], coord_h[1]]))
        elif f([coord_h[0], coord_h[1] - h]) < f(coord_h):
            while f([coord_h[0], coord_h[1] - h]) < f(coord_h):
                coord_h[1] -= h
                
        print("Значение координат: ", coord_h[0], coord_h[1],"Значение целевой функции: ",  f([coord_h[0], coord_h[1]]))
        return coord_h[1]
```

Метод покоординатного спуска:

```python
def coordinate_descend(f, coord_h):
    
    eps = 0.001
    h = 2
    coord_h_new = coord_h
    for i in range(5):
        coord_h_old = list(coord_h_new[:])
        #print(minimization(f, coord_h_old, h, True))
        coord_h_new[0] = minimization(f, coord_h_old, h, True)
        #print(minimization(f, coord_h_new, h, False))
        coord_h_new[1] = minimization(f, coord_h_new, h, False)
        
        h /=2
        if (f([coord_h_new[0] - h,coord_h_new[1]]) < f(coord_h_new)) and (f([coord_h_new[0] + h,coord_h_new[1]]) < f(coord_h_new)) and (f([coord_h_new[0], coord_h_new[1]- h]) < f(coord_h_new)) and (f([coord_h_new[0], coord_h_new[1]+h]) < f(coord_h_new)) :
            break
    return coord_h_new
```
