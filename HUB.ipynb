# Вычисление геодезического расстояния

```python
def geodetic_range(coord_1, coord_2):
    
    R = 6371.3
    a1 = coord_1[0] * math.pi / 180
    a2 = coord_2[0] * math.pi / 180
    b1 = coord_1[1] * math.pi / 180
    b2 = coord_2[1] * math.pi / 180
    
    x1 = R * math.cos(a1) * math.cos(b1)
    y1 = R * math.cos(a1) * math.sin(b1)
    z1 = R * math.sin(a1)
    
    x2 = R * math.cos(a2) * math.cos(b2);
    y2 = R * math.cos(a2) * math.sin(b2);
    z2 = R * math.sin(a2);
    
    dist = pow(pow((x2 - x1), 2) + pow((y2 - y1), 2) + pow((z2 - z1), 2), 1 / 2)
    alfa = 2 * math.asin(dist / (2 * R))

    return alfa
```

Целевая функция:

```python
def func(coord_h):
    
    sum = 0
    for i in range(0, 71):
        sum += math.exp(geodetic_range(eu_df['airport coordinates'][i], coord_h)) * eu_df['passenger traffic'][i]
    
    for i in range(0, 50):
        sum += math.exp(geodetic_range(aus_df['airport coordinates'][i], coord_h) * aus_df['passenger traffic'][i])
    
    return sum
```

Задание отрезков для поиска оптимума методом золотого сечения и начальной точки для координатного спуска:

```python
lon_for_golden_ratio = [20.0, 127.0] #отрезок для долготы [26.4, 70] 13.197889
lat_for_golden_ratio = [6.0, 60.0] #отрезок для широты [20, 43] 77.706333
prosp_h = [13.0, 77.0] #предполагаемое положение хаба  6°44' и 35°30' северной широты и 68°7' и 97°25' восточной 46635095282
#1179509999998
```

Метод золотого сечения для одномерной оптимизации:

```python
t =  0.5 * (1.0 + (5.0)**0.5)

def minimization(f, a, b, n, coord_h):
    
    eps = 0.000001
    if abs(a - b) < eps:
        return 0.5 * (a + b)
    else:
        step = (b-a)/t
        B,A = b - step, a + step
        if n:
            if f([B, coord_h[1]]) >= f([A, coord_h[1]]): 
                 return minimization(f, B, b, n, coord_h)
            else:
                return minimization(f, a, A, n, coord_h)
        else:
            if f([coord_h[0], B]) >= f([coord_h[0], A]): 
                 return minimization(f, B, b, n, coord_h)
            else:
                return minimization(f, a, A, n, coord_h)
```

Метод покоординатного спуска:

```python
def coordinate_descend(f, coord_h):
    
    eps = 0.000001
    coord_h_new = coord_h
    for i in range(10):
        coord_h_old = list(coord_h_new[:])
        coord_h_new[0] = minimization(f, lat_for_golden_ratio[0], lat_for_golden_ratio[1], True, coord_h_old)
        coord_h_new[1] = minimization(f, lon_for_golden_ratio[0], lon_for_golden_ratio[1], False, coord_h_new)
        
        if abs(f(coord_h_new) - f(coord_h_old)) < eps :
            break
    return coord_h_new
```

Координаты хаба в десятичных градусах широты и долготы:

```python
coordinate_descend(func, prosp_h)
```
